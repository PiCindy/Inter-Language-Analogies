```python
import nlg
import random
from nlg.Vector.Vectors import Vectors
```

```python
def create_vectors(file):
    sigmorphon_lines = [line.strip() for line in open(file)]
        
    char_feature = True
    morph_feature = True
    lemma_feature = False
    gen_lemma = False

    sigmorphon_vectors = Vectors.fromSigmorphonFile(lines=sigmorphon_lines,
                                char_feature=char_feature,
                                morph_feature=morph_feature,
                                lemma_feature=lemma_feature,
                                gen_lemma=gen_lemma)
    
    return sigmorphon_vectors
```

```python
def create_file(file, vectors):
    open(file, 'w').close()
    with open(file, 'w') as f:
        for vector in vectors:
            f.write(vector)
            f.write('\n')
```

```python
def create_shuffled_file(file, vectors):
    random.shuffle(vectors)
    print(vectors)
    return create_file(file, vectors)
```

```python
sig_file = 'data/ceb.hil'
vec_file = 'data/ceb.hil.vec2'
vectors = create_vectors(sig_file)
create_shuffled_file(vec_file, vectors)
!python3 Lines2Vectors.py <'data/ceb.hil.vec2' -V | python3 Vectors2Grids.py -V > 'data/ceb.hil.grid2'
```

```python
def create_files(file, vectors1, vectors2):
    open(file, 'w').close()
    with open(file, 'w') as f:
        for vector in vectors1:
            f.write(vector)
            f.write('\n')
        for vector in vectors2:
            f.write(vector)
            f.write('\n')
```
